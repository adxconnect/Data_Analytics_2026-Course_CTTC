```python
import pandas as pd

# Reading Dataset
# The iris.csv from seaborn-data typically has a header, so we let pandas infer it.
iris = pd.read_csv(r"D:\Data_Analytics_2026(Course)\Python\Dataset\iris\iris.data", header=None, names=['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'species'])
iris
```

```python
sns.distplot(iris.sepal_length)
plt.show()
```

```python
sns.histplot(iris.sepal_length)
plt.show()
```

```python
sns.kdeplot(iris.sepal_length)
plt.show()
```

```python
sns.histplot(iris.sepal_length, fill=False,color='orange',hatch="o")
plt.show()
```

```python
sns.kdeplot(iris.sepal_length,fill=True)
plt.show()
```

```python
sns.kdeplot(data=iris[iris['species']== 'setosa'], x='sepal_length', fill=True, label='setosa')
sns.kdeplot(data=iris[iris['species'] == 'versicolor'], x='sepal_length', fill=True, label='versicolor')
sns.kdeplot(data=iris[iris['species'] == 'virginica'], x='sepal_length', fill=True, label='virginica')
plt.title('Distribution of Sepal Length by Species (KDEplot)')
plt.legend()
plt.show()
```

```python
sns.histplot(iris.sepal_length[iris['species']== 'setosa'], color='blue', label='setosa')
sns.histplot(iris.sepal_length[iris['species'] == 'versicolor'], color='green', label='versicolor')
sns.histplot(iris.sepal_length[iris['species'] == 'virginica'], color='red', label='virginica')
plt.title('Distribution of Sepal Length by Species (Histplot)')
plt.legend()
plt.show()
```

```python
sns.kdeplot(data=iris[iris['species']== 'setosa'], x='sepal_length', fill=True, label='setosa')
sns.kdeplot(data=iris[iris['species'] == 'versicolor'], x='sepal_length', fill=True, label='versicolor')
sns.kdeplot(data=iris[iris['species'] == 'virginica'], x='sepal_length', fill=True, label='virginica')
plt.title('Distribution of Sepal Length by Species (KDEplot)')
plt.legend()
plt.show()
```

```python
sns.kdeplot(data=iris,fill=True) # This plot is commented out as it does not specify an x or y variable
plt.show()
```
