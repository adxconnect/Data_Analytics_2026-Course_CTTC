```python
# The dataset loaded with header already has appropriate column names.
# If you still want to rename them for consistency with earlier attempts:
iris.columns = ['SL','SW','PL','PW','Flower Type']

# Ensure numerical columns are of numeric type, coercing errors to NaN
iris['SL'] = pd.to_numeric(iris['SL'], errors='coerce')
iris['SW'] = pd.to_numeric(iris['SW'], errors='coerce')
iris['PL'] = pd.to_numeric(iris['PL'], errors='coerce')
iris['PW'] = pd.to_numeric(iris['PW'], errors='coerce')

# Drop any rows where these conversions resulted in NaN values
iris.dropna(subset=['SL', 'SW', 'PL', 'PW'], inplace=True)

iris.info() # Add this to verify dtypes
iris
```
