```python
import pandas as pd
import numpy as np

# Load the dataset again to ensure 'auto' is defined correctly as a DataFrame
auto = pd.read_csv(r"d:/Data_Analytics_2026(Course)/Python/Dataset/auto+mpg/auto-mpg.data", header=None, sep=r'\s+')
auto.columns = ['mpg', 'cylinders', 'displacement', 'horsepower', 'weight', 'acceleration', 'model year', 'origin', 'car name']

# Convert 'horsepower' to numeric, forcing '?' to NaN
auto['horsepower'] = pd.to_numeric(auto['horsepower'], errors='coerce')

# Fill NaN values in 'horsepower' with the mean of the column
auto['horsepower'] = auto['horsepower'].fillna(auto['horsepower'].mean())
```

**Matplotlib**
<pre> <b> Data Visulaization Technique
Matplotlib is a python library for data visdulization, primarily used to create static, animated, interactive plots.
It provides a wide range of plotting functions to visualize data effectively.

```python
import matplotlib.pyplot as plt
```

```python
#line plot
x = [1,9,6,4,6,10,4,6,13]
y = [9,6,8,4,5,3,2,11,18]
plt.plot(x,y,"sg-.")
plt.title('Line Plot',color='purple')
plt.xlabel('X-Axis',color='r')
plt.ylabel('Y-Axis',color='y')
plt.grid()
plt.show()
```

```python
#Bar Plot
x=[1,3,5,7,9]
y=[2,4,6,8,10]
plt.bar(x,y,color=['b','g','r','y','m'],width=1.0)
plt.title("Bar Plot")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.annotate("Happy",xy=(0.5,2.5))
plt.show()
```

```python
#Multiple Bar Plot
x=[1,3,5,7,9]
y=[2,4,6,8,10]
plt.bar(x,y,label='Example 1')
plt.show()
x1=[4,7,2,9,8]
y1=[9,2,6,4,1]
plt.bar(x1,y1,label='Example 2')
plt.show()
plt.title("Bar Plot")
plt.legend(["A","B"])
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.show()
```

```python
#Histogram
ages=[35,10,2,8,5,15,20,28,24,27,45,30]
bins=[0,10,20,30,40,50]
plt.hist(ages,bins,histtype='bar',rwidth=0.8,edgecolor='k',linewidth=3)
plt.title("Histogram ")
plt.xlabel("Ages")
plt.ylabel("Value")
plt.show()
```

```python
#Scatter Plot
x=[3,5,4,7,2,10,6,8]
y=[2,6,4,9,7,9,3,10]
plt.scatter(x,y,s=300,color='Gold',marker='*',edgecolors='orange',linewidths=10)
plt.scatter(x,y,color='brown')
plt.bar(x,y,color='g',width=0.05)
plt.title("Scatter Plot")
plt.show()
```

```python
#Pie Chart
slices=[8,6,10,4,2]
activities=['Sona','Padhai','Phone','Khana','Game']
cols=['r','g','gold','c','m']
plt.pie(slices,labels=activities,colors=cols,startangle=90,shadow=True,explode=(0,0.1,0,0,0),autopct="%1.2f%%")
plt.title("Pie Chart")
plt.show()
```
