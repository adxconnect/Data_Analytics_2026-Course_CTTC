**Machine Learning** is a subset of AI that enables machine to learn the patterns from data and make predictions without being explicitly program.It can extract features and patterns from the data.

It is used for making predictive models and can solve complex problems.

Ex - Auto Fill on Search Engine, Product Recommendation, Google Translation, Youtube Recommendations, Traffic Alert using Google Maps,etc

Machine Learning is of three types :

1> Supervised Learning
2> Unsupervised Learning
3> Reinforcement Learning

**1. Supervised Learning** - In this machine learns under supervision. It is used for labelled data. It is a type of machine learning where the model is trained using label data. It is divided into parts - Regression (Op:Continuous) and Classification. and Classification (Op: Categorical).

For Regression -
1.  Linear Regression
2. Polynomial
3. Lasso
4. Ridge

For Classification
(Op: Categorical)
1. Logistic Regression
2. Naive Bayes
3. KNN (Nearest Neighbors)
4. SVM (support vector machine)
5. Decision Tree
